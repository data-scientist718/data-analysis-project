## Now we visualize the data with analysis

```python
# Show Show Percentage of All Employees per Age Group 

plt.figure(figsize = (5,5))


hr_analysis['age_group'].value_counts().plot(kind='pie', title = "Percentage of All Employees Per Age Group");
```

```python
#Show Percentage of Attrition Per Age Group

plt.figure(figsize = (5,5))

hr_analysis_attrition['age_group'].value_counts().plot(kind='pie', title = "Percentage of Attrition Per Age Group");
```

Millennials comprise the largest age group employed by the company. This group also represents the highest level of attrition. milenium age is (28 to 46)

```python
# check the attrition level of employee in educationfield group by gender
plt.figure(figsize = (10,5))

plt.title("Attrition by Gender and Education Field")
sns.countplot(x = 'Gender', hue = 'EducationField', data = hr_analysis_attrition);
```

The attrition level is higher among male employees. Employees with education in life sciences and medicine leave the company at a higer rate than other professionals.

```python
plt.figure (figsize=(10,6))

plt.title("attrition by gender and Job role")

sns.countplot( x ="Gender" , hue= "JobRole" , data= hr_analysis_attrition);
```

Attrition levels are highest among men performing jobs as Sales Executives, Laboratory Technicians and Research Scientists. For women, attrition is highest at the Sales Executive level. Men and women are less likely to quit the Research Director or Manager job role.

```python
# This code creates a pie chart showing the attrition of employees by their environment satisfaction level.

plt.figure(figsize = (10,10))

plt.subplot(1,2,1)

hr_analysis_attrition['EnvironmentSatisfaction'].value_counts().plot(kind='pie', title = "Attrition by Environment Satisfaction")

# This code creates a pie chart showing the attrition of employees by their Work Life Balance level.

plt.subplot(1,2,2)

hr_analysis_attrition['WorkLifeBalance'].value_counts().plot(kind='pie', title = "Attrition by Work Life Balance");
```

On a scale between 1 - 4 with four being the highest level of Environment Satisfaction, attrition was equal between levels 3 and 4. On a scale of 1 - 4 with four being the highest level of Work Life balance more than half of attritioned employees viewed this category as level 3. Only a small percentage of employees reporting optimal Work life Balance left the company.

```python
# now we  want to get attrition by maritaul status

hr_analysis_attrition["MaritalStatus"].value_counts().plot(kind = "pie" , title = "Attrition by Marital Status");
```

Married employees attritioned at a higher level than single or divorced employees. 

```python
# now we  want to get attrition by job level

hr_analysis_attrition["JobLevel"].value_counts().plot(kind = "pie" , title = "Attrition by job levael");
```

 Employees unhappy with their job level are more likely to attrition than others.

```python
# now we want to get the Attrition by Marital Status of Gender
plt.figure(figsize = (10,5))

plt.title("Attrition by Gender Marital Status")
sns.countplot(x = 'Gender', hue = 'MaritalStatus', data = hr_analysis_attrition);
```

Single men account for a high propotion of attrition followed by married men. Divorced women account for the lowest level of attrition.

```python
# Find Correlation of Monthly Income and Total Working Years

plt.figure(figsize=(10,6))

plt.title("Correlation of Monthly Income and Total Working Years")

sns.scatterplot(x=hr_analysis_attrition['TotalWorkingYears'], y=hr_analysis_attrition['MonthlyIncome']);
```

 we see in this  scatter plot As number of years employeed by the company increases, monthly income increases.

```python
# find The attrition in eductaion group by jobrole
plt.figure(figsize = (10,5))

plt.title("Attrition by Education and Job Role")
sns.countplot(x = 'Education', hue = 'JobRole', data = hr_analysis_attrition);
```

When education combined with job roles are considered, employees who have obtained education at level 3 are more likely to leave the company. Within that group, Laboratory Technicians are at the greatest risk of leaving, followed by Sales Executives.

```python
# Find the Attrition of bussinees travel group by  environment satisfaction

plt.figure(figsize = (10,5))

plt.title("Attrition by Business Travel and Environment Satisfaction")
sns.countplot(x = 'BusinessTravel', hue = 'EnvironmentSatisfaction', data = hr_analysis_attrition);
```

Overall, employees that travel rarely leave at a higher level than employees that travel frequently or those that don't travel at all. Employees that travel rarely are less satified with work environment and account for the highest level of attrition.

```python
plt.figure(figsize = (10,5))

plt.title("Attrition by Number of Companies Worked and Age Group")
sns.countplot(x = 'NumCompaniesWorked', hue = 'age_group', data = hr_analysis_attrition);
```

Gen Z and Millennials are most likely to leave if they have only worked for 1 company prior to employment.

```python
plt.figure(figsize = (14,5))

plt.title("Attrition by Years Since Last Promotion and Age Group")
sns.countplot(x = 'YearsSinceLastPromotion', hue = 'age_group', data = hr_analysis_attrition);
```

Millennials are most likely to leave within one year of not getting a promotion.

```python
plt.figure(figsize = (10,5))

plt.title("Attrition by Years in Current Role and Age Group")
sns.countplot(x = 'YearsInCurrentRole', hue = 'age_group', data = hr_analysis_attrition);
```

Gen Z employees and Millennials are most likely to leave before completing an entire year in a job role. However, Millenials are most likely to leave within the second year of employment.

```python
# Show Attrition by Stock Option Level

plt.figure(figsize = (10,10))

plt.subplot(1,2,1)

hr_analysis_attrition['StockOptionLevel'].value_counts().plot(kind='pie', title = "Attrition by Stock Option Level")

#Show Attrition by Job Level

plt.subplot(1,2,2)

hr_analysis_attrition['JobLevel'].value_counts().plot(kind='pie', title = "Attrition by Job Level");
```

Employees with the highest stock option level,2 are less likely to leave the company. Additionally, employees with the highest job level, 5 are less likely to leave the company.

```python
# Show Attrition by Raise and Performance Rating
plt.figure(figsize = (10,5))


plt.title("Attrition by Raise and Performance Rating")
sns.countplot(x = 'PercentSalaryHike', hue = 'PerformanceRating', data = hr_analysis_attrition);
```

Employees receiving a Performance Rating of 3 are more likely to leave the company. Very few employees leave that have the highest Performance Rating of 4 with a comparable 25% pay raise.

```python
# Show Attrition by Age and Distance from Home
plt.figure(figsize = (15,7))


plt.title("Attrition by Age Group and Distance from Home ")
sns.countplot(x = 'DistanceFromHome', hue = 'age_group', data = hr_analysis_attrition);
```

Distance from home to work, when measured from 1 mile up to 29 miles, indicates Millenials are most likely to leave even if the distance from work to home is as low as 1 mile.
