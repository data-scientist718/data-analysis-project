```python
# now we calculate the each category total sales in from high to low

sales_by_category = Data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
sales_by_category
```

```python
#now we show the above data in vertical bar charts

sales_by_category.plot(kind='bar', color='green')
plt.show()
```

```python
# now we want to check the average sale of month

Data['Order Date'] = pd.to_daStetime(Data['Order Date'])
Data['Month'] = Data['Order Date'].dt.month
average_sales_per_month = Data.groupby('Month')['Sales'].mean()
average_sales_per_month
```

```python
Data.info()
```

```python
# now we show the line chart of average monthly sales.

average_sales_per_month.plot(kind='line', color='red', ls='--')
plt.show()
```

```python
# now we want to get sales by states

sales_by_state = Data.groupby('State')['Sales'].sum().sort_values(ascending=False)
sales_by_state.plot(kind = "bar" , color = "red")
plt.show()
```

```python
# creating a horizontal bar chart to show the total sales of products

sales_by_product = Data.groupby('Product Name')['Sales'].sum().sort_values(ascending=True)
sales_by_product.head(7).plot(kind='barh', color='yellow')
plt.show()
```

```python
#creates a scatter plot that shows the total sales for the top 20 customers with the highest sales in the Data dataframe.

sales_by_customer = Data.groupby('Customer Name')['Sales'].sum()
plt.scatter(sales_by_customer.head(20).index, sales_by_customer.head(20).values, c ='red')
plt.xlabel('Customer')
plt.xticks(rotation=90)
plt.ylabel('Sales')
plt.show()
```

```python
# Now we find the monthly sales growth percentage over time.

monthly_sales = Data.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum()
monthly_sales_growth = monthly_sales.pct_change()

monthly_sales_growth
```

```python
# Now we creating a line plot to visualize the monthly sales growth percentage over time.

monthly_sales_growth.plot(kind='line', ls='-', color='orange')
plt.xlabel('Month')
plt.ylabel('Sales Growth')
plt.show()
```

```python
sales_by_month_category = Data.groupby(['Month', 'Category'])['Sales'].sum().unstack()
sales_by_month_category
```

```python
#this code is a useful way to visualize sales trends across different categories over time. 
#The resulting heatmap allows us to easily identify which categories had the highest sales in each month,

plt.pcolor(sales_by_month_category, cmap='Blues')
plt.xlabel('Category')
plt.ylabel('Month')
plt.xticks(rotation=45)
plt.colorbar()
plt.show()
```
